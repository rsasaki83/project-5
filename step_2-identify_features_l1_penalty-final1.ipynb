```python
madelon_feat_df= load_data_from_database()
```

```python
X=madelon_feat_df.drop(['index', 'label'], axis=1)
y=madelon_feat_df['label']
```

```python
make_data_dict(X,y)
```

```python
data_dict=make_data_dict(X,y)
```

```python
from sklearn.preprocessing import StandardScaler

general_transformer(StandardScaler(),data_dict)
```

```python
data_dictionary=general_transformer(StandardScaler(),data_dict)
```

```python
from sklearn.linear_model import LogisticRegression
general_model(LogisticRegression(penalty='l1'), data_dictionary)
```

```python
final_data_dict=general_model(LogisticRegression(penalty='l1'), data_dictionary)
```

```python
final_data_dict['train_score'],final_data_dict['test_score']
```

```python
coef1=final_data_dict['model']
coef1.coef_
```

```python
import pandas as pd
Positive_results=pd.DataFrame(coef1.predict_proba(X))
Positive_results.head()
```
